## Tehtävä 1

```python
import numpy as np

A0 = np.array([[3], [2], [5]])
W1 = np.array([[1.1, -1.3, 1.5], [0.1, 2.0, 1.7]])
W2 = np.array([[2.0, -1.2]])

A1 = np.dot(W1, A0)
A2 = np.dot(W2, A1)
```

```python
print("Matriisi A1:")
print(A1)
print("Matriisi A2:")
print(A2)
```

## Tehtävä 2

```python
def relu(x):
    return np.maximum(x, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

A1 = relu(np.dot(W1, A0))
A2 = sigmoid(np.dot(W2, A1))
```

```python
print("Matriisi A1:")
print(A1)
print("Matriisi A2:")
print(A2)
```

## Tehtävä 3

```python
vector_x = np.arange(1, 11).reshape(-1, 1) # luo sarjan 1-10 ja muuttaa sen vektoriksi
vector_y = np.array([[14], [12], [34], [24], [43], [38], [41], [57], [55], [50]])

print("Vektori x:")
print(vector_x)
print("Vektori y:")
print(vector_y)
```

```python
ones = np.ones(10)
X = np.vstack([ones, vector_x.reshape(1, -1)])
X = X.T

print("Matriisi X:")
print(X)
print(vector_y.reshape(1, -1))
```

```python
M = np.linalg.inv(X.T @ X) @ X.T @ vector_y

print("Matriisi M:")
print(M)
```

```python
import matplotlib.pyplot as plt

xx = np.linspace(1, 10, 1000)
XX = np.column_stack((np.ones_like(xx), xx))
yy = XX.dot(M)
plt.plot(vector_x, vector_y, 'o')
plt.plot(xx, yy)
```
